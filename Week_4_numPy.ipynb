"""
    Extract unique chair professor names (Prof./Professor) from HTML text.
    Supports Unicode, apostrophes, accents, initials, and is case-insensitive.
    """
    # This pattern:
    # - (?i): case-insensitive
    # - (Prof\.|Professor): Prof. OR Professor (any case)
    # - \s+: at least one space (can extend to optional if desired)
    # - Name: starts with capital, includes letters, hyphens, apostrophes, accents, dots, and spaces
    # - Looks ahead for a word boundary or punctuation
    pattern = r'(?i)(Prof\.|Professor)\s+([A-Z][\w\'\-\.\sÀ-ÖØ-öø-ÿ]*[A-Za-zÀ-ÖØ-öø-ÿ])\b'
    matches = re.findall(pattern, html, flags=re.UNICODE)
    names = []
    seen = set()
    for prefix, name in matches:
        # Remove accidental trailing punctuation
        name = re.sub(r'[\.,;:]+$', '', name.strip())
        full_name = f"{prefix} {name}"
        if full_name not in seen:
            names.append(full_name)
            seen.add(full_name)
    return names

```python
import re

html_content = """

<p><table border="1" cellspacing="0" cellpadding="3">

<tbody>

<tr>

<th style="text-align: left; background-color: #ccc1d9;" width="23%">Year</th>

<th style="text-align: left; background-color: #ccc1d9;" width="77%">History</th>

</tr>

<tr class="odd">

<td>1979</td>

<td>Statistics (STAT) Programme was set up as an individual programme of study.</td>

</tr>

<tr class="odd">

<td>1981</td>

<td>Establishment of Division of Statistics at Graduate School, CUHK.</td>

</tr>

<tr class="odd">

<td>1982</td>

<td>Foundation of the Department. 17 students were enrolled in the first year with Prof. Howell Tong as the chairman</td>

</tr>

<tr class="odd">

<td>1986</td>

<td>Professor Sik-yum Lee assumed the duty of chairman of the Department.</td>

</tr>

<tr class="odd">

<td>1993</td>

<td>Professor Nai Ng Chan became the third chairman of the Department.</td>

</tr>

<tr class="odd">

<td>1994</td>

<td>The Department office relocated from Science Centre to the current site &#8211; Lady Shaw Building.</td>

</tr>

<tr class="odd">

<td></td>

<td>Professor Wing-hung Wong served as the chairman of the Department.</td>

</tr>

<tr class="odd">

<td>1998</td>

<td>Professor Yeh Lam succeeded Professor Wing-hung Wong and took up the chairmanship of the Department.</td>

</tr>

<tr class="odd">

<td>1999</td>

<td>Risk Management Science (RMSC) stream was launched. Professor Kim Hung Li assumed the chairmanship for one year.</td>

</tr>

<tr class="odd">

<td>2000</td>

<td>Professor Jian Qing Fan served as the seventh chairman of the Department.</td>

</tr>

<tr class="odd">

<td></td>

<td>Risk Management Science (RMSC) was expanded to be an individual programme of study.</td>

</tr>

<tr class="odd">

<td></td>

<td>The first batch of students majoring in RMSC was enrolled.</td>

</tr>

<tr class="odd">

<td>2004</td>

<td>Professor Ngai Hang Chan was inaugurated as the Department chairman.</td>

</tr>

<tr class="odd">

<td>2009</td>

<td>The Department first offered a double-major programme in Quantitive Finance and Risk Management Science (QRFM).</td>

</tr>

<tr class="odd">

<td>2012</td>

<td>Corresponding to the new 3-3-4 curriculum, the undergraduate programme in Statistics joined Science Faculty broad-based admission scheme.</td>

</tr>

<tr class="odd">

<td>2013</td>

<td>Professor Qi Man Shao was inaugurated as the Department chairman.</td>

</tr>

<tr class="odd">

<td>2018</td>

<td>Professor Xinyuan Song was inaugurated as the Department chair.</td>

</tr>

<tr class="odd">

<td>2022</td>

<td>B.Sc. in Computational Data Science was launched.</td>

</tr>

<tr class="odd">

<td>2024</td>

<td>M.Sc. in Advanced Studies in Statistics and Data Science was launched.</td>

</tr>

<tr class="odd">

<td>2024</td>

<td>Professor Junhui Wang was inaugurated as the Department chair.</td>

</tr>

</tbody>

</table>

</p>

"""

def decode_html_entities(text):
    # Covers key HTML entities found in your sample and some common extras
    text = text.replace('&#8211;', '–').replace('&ndash;', '–')
    text = text.replace('&nbsp;', ' ')
    text = text.replace('&amp;', '&')
    # Add more as needed for your HTML
    return text

def extract_chair_professors(html_content):
    # Step 1: Decode known HTML entities
    text_content = decode_html_entities(html_content)
    # Step 2: Remove HTML tags (robust for simple HTML)
    text_content = re.sub(r'<[^>]+>', ' ', text_content)

    # Step 3: Regex for professor names (now supports initials, hyphens, apostrophes, dots, and non-ASCII letters)
    # - Names: sequences of capitalized words, initials with dots, hyphens, apostrophes, and Unicode letters
    # - At least two "name parts" after the title
    name_part = r"(?:[A-ZÀ-ÖØ-öø-ÿĀ-ž][a-zA-ZÀ-ÖØ-öø-ÿĀ-ž'\-]*|[A-Z]\.)"
    professor_pattern = (
        r'\b(?:Prof\.|Professor)\s+' +
        name_part + r'(?:\s+' + name_part + r'){1,4}'
    )
    matches = re.findall(professor_pattern, text_content)

    # Step 4: Clean up whitespace and deduplicate while preserving order
    cleaned = []
    seen = set()
    for m in matches:
        name = ' '.join(m.split())
        if name not in seen:
            cleaned.append(name)
            seen.add(name)
    return cleaned

if __name__ == "__main__":
    professors = extract_chair_professors(html_content)
    for prof in professors:
        print(prof)
```

# Numpy as Matrix Calculator

```python
# arithmetics (element-wise)
x = np.array([1,2,3])
y = np.array([4,5,6])
print(x+y, x-y, x*y, x/y, x**2)
```

```python
x.T
```

```python
print(x.dot(y))
print(np.dot(x,y))
```

```python
# Inner / Dot Product
x.dot(y) # (x^T)y
np.inner(x, y)
```

```python
# Outer Product
np.outer(x, y)
```

```python
# transpose
z = np.array([y, y**2])
print(z, z.shape)
print(z.T, z.T.shape)
```

```python
# matrix multiplication
print(np.matmul(z, z.T))
print(z.dot(z.T))
```

```python
# inverse

from numpy.linalg import inv
a = np.array([[1., 2.], [3., 4.]])
ainv = inv(a)
print(ainv)
```

```python
# check AA^(-1) = I
print(np.allclose(np.dot(a, ainv), np.eye(2)))
```

```python
np.dot(a, ainv)== np.eye(2)
```

# Matrix Slicing

```python
# Array
s = np.arange(13)**2
print(s[0], s[4], s[-1], s[1:5], s[-4:])
print(s[-5::-2]) # array[start:stop:stepsize]
```

```python
# Matrix
r = np.arange(36)
r.resize((6, 6))
print(r)
print(r[2, 2], r[3, 3:6], r[:2, :-1], r[-1, ::2], r[r > 30])
r[r > 30] = 30
print(r)
```
